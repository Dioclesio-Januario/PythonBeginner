## ## Configuração da matplotlib
- O comportamento default pode ser personalizado por meio de um conjunto amplo de parâmetros globais que determinam o tamanho da figura, o espaçamento das subplotagens, as cores, os tamanhos das fontes, os estilos de grade e assim por diante.
- O primeiro argumento de rc é o componente que desejamos personalizar, como: 'figure', 'axes', 'xtick', 'ytick', 'grid', 'legend' ou vários outros.
- Depois disso, pode haver uma sequência de argumentos nomeados informando os novos parâmetros.

```python
# Definir o tamanho default da figura globalmente.
dados = np.arange(11)
plt.plot(dados)

plt.rc('figure', figsize = (6, 4))

plt.show()
```

## Figuras e subplotagens

```python
# Vários gráficos na mesma figura.

plt.subplot(2, 2, 1)
dados1 = np.random.randn(50).cumsum()
plt.plot(dados1, 'k--')                                # k é o estilo do tracejado.

plt.subplot(2, 2, 2)
dados2 = np.random.randn(100)
plt.hist(dados2, bins = 20, color='k', alpha= 0.5)     # alpha é a opacidade; bins é a repartição do hist.

plt.subplot(2, 2, 3)
plt.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))

plt.subplot(2, 2, 4)
dados4 = np.arange(30)
plt.plot(dados4, 'k--')

plt.show()
```

```python
# Ajustando a escala dos boxes.
plt.subplot(2, 2, 1)
dados1 = np.random.randn(50).cumsum()
plt.plot(dados1, 'k--')                                # k é o estilo do tracejado.

plt.subplot(2, 2, 2)
dados2 = np.random.randn(100)
plt.hist(dados2, bins = 20, color='k', alpha= 0.5)     # alpha é a opacidade; bins é a repartição do hist.

plt.subplot(2, 2, 3)
plt.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))

plt.subplot(2, 2, 4)
dados4 = np.arange(30)
plt.plot(dados4, 'k--')

plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
plt.show()
```

```python
# Orientada a objecto.
fig, ax = plt.subplots(2, 2)

dados1 = np.random.randn(50).cumsum()
dados2 = np.random.randn(100)
dados4 = np.arange(30)

ax[0, 0].plot(dados1, 'k--')                                # k é o estilo do tracejado.
ax[0, 1].hist(dados2, bins = 20, color='k', alpha= 0.5)     # alpha é a opacidade; bins é a repartição do hist.
ax[1, 0].scatter(np.arange(30), np.arange(30) + 3*np.random.randn(30))
ax[1, 1].plot(dados4, 'k--')

plt.show()
```

## Personalização
- Cor: preto (k); verde (g); vermelho (r); azul (b); amarelo (y)...
- Estilo de linha: -; --; .; -.; +; *; ...
- Marcador: .; o; x; *; ^; s; w+; ...
- Espessura da linha: lw = 2.

```python
# Modo explícito.
plt.plot(np.arange(30), color = 'r', linestyle = '-', marker = '.')
plt.show()
```

```python
# Modo implícito: Cor - marcador - estilo de linha
plt.plot(np.arange(30), 'y-.')
plt.show()
```

```python
from numpy.random import randn
plt.plot(randn(30).cumsum(), 'ko--')
plt.show()
```

```python
# Grid
from numpy.random import randn
plt.plot(randn(30).cumsum(), 'ko--')
plt.grid()
plt.show()
```

```python
# Espessura, anotações e limites.
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)

plt.plot(t, s, lw = 1)                                          # lw = espessura da linha.
plt.annotate('max', xy = (2, 1), xytext = (3, 1.5),
arrowprops = dict(facecolor='black', shrink= 0))

plt.ylim(-2, 2)
plt.show()
```

```python
# Fundo preto
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'w-.')
plt.show()
```

## Título, Tiques, rótulos e legendas

```python
# Título.
dados = np.random.randn(1000).cumsum()
plt.plot(dados)
plt.title('My first matplotlib plot')
plt.show()
```

```python
# Rótulos.
dados = np.random.randn(1000).cumsum()

plt.plot(dados)
plt.title('My first matplotlib plot')

plt.xlabel('Stages')
plt.ylabel('y')
plt.show()
```

```python
# Legendas.
plt.plot(np.random.randn(100).cumsum(), 'k', label = 'one')
plt.plot(randn(100).cumsum(), 'k--', label = 'two')
plt.plot(randn(100).cumsum(), 'k.', label = 'three')

plt.legend(loc='best')

# Ou
# plt.legend(loc="lower right", title="Legend Title", frameon=False)

plt.show()
```

```python
x = np.linspace(0, 1, 100)

# (x, y, legenda)
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

# Eixos
plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot([1, 2, 3, 4])

plt.title("Simple Plot")
plt.xlabel('x label')
plt.ylabel('y label')

plt.grid()
plt.show()
```

## Tipos de gráficos

```python
names = ['group_a', 'group_b', 'group_c']
values = [10, 20, 50]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)

plt.subplot(132)
plt.scatter(names, values)

plt.subplot(133)
plt.plot(names, values)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
# Barras horizontal.
names = ['group_a', 'group_b', 'group_c']
values = [10, 20, 50]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.barh(names, values)
plt.show()
```
